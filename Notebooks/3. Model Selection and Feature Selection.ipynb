```python
# importing necessary libraries

import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import MinMaxScaler

# importing regression models

from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
```

```python
home_df= pd.read_csv("../dataframes/mapped_df.csv")
```

```python
home_df.head()
```

```python
# removing unnecessary columns

home_df.drop('Property_ID',axis=1,inplace=True)
```

```python
# Seperating the target column

y=home_df.pop('Habitability_score')
X = home_df
```

```python
# Scaling the data

normal = MinMaxScaler()
Scale_X=normal.fit_transform(X)
```

#### Scaling data only used by linear model, Elasticnet model and SVR model

```python
# Checking cross validation score of each model

lr_score = cross_val_score(LinearRegression(),Scale_X,y, scoring='r2', cv=3)
en_score = cross_val_score(ElasticNet(),Scale_X,y, scoring='r2', cv=3)
svr_score = cross_val_score(SVR(),Scale_X,y, scoring='r2', cv=3)
rf_score = cross_val_score(RandomForestRegressor(n_jobs=-1),X,y, scoring='r2', cv=3)
ada_score = cross_val_score(AdaBoostRegressor(),X,y, scoring='r2', cv=3)
gb_score = cross_val_score(GradientBoostingRegressor(),X,y, scoring='r2', cv=3)
xg_score = cross_val_score(XGBRegressor(),X,y, scoring='r2', cv=3)
cat_score = cross_val_score(CatBoostRegressor(),X,y, scoring='r2', cv=3)
```

```python
print("Linear Regression: ", lr_score.mean())
print("ElasticNet Regression: ", en_score.mean())
print("SVR ", svr_score.mean())
print("RandomForest Regression: ", rf_score.mean())
print("AdaBoost Regression: ", ada_score.mean())
print("GradientBoost Regression: ", gb_score.mean())
print("XGB Regression: ", xg_score.mean())
print("Catboost Regressor: ", cat_score.mean())
```

#### Top 3 models :- Random Forest , CatBoost , XBG boost

## Feature Selection

```python
# mutual_info_regression function used to find information gain of each columns 

from sklearn.feature_selection import mutual_info_regression
```

```python
info = mutual_info_regression(X,y)
info
```

```python
# mapped information with their column name

col_info = list(zip(X.columns,info))
col_info
```

```python
# Sort columns according to the more information gain

col_info = sorted(col_info,reverse=True,key=lambda x:x[1])
col_info, len(col_info)
```

```python
#selecting only those columns with information gain>0.015 

selected_cols = [ci[0] for ci in col_info if ci[1]>0.015]
selected_cols
```

```python
New_X = X[selected_cols]
New_X.head()
```

```python
df2 = pd.concat([New_X,y],axis=1)
```

#### At the end, we have best models and best features

## Saving

```python
df2.to_csv(r'../dataframes/model_data.csv',index=False)
```
