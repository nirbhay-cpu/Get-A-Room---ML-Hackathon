```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb


from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import r2_score

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
home_df= pd.read_csv("../dataframes/model_data.csv")
```

```python
home_df.head()
```

```python
y=home_df.pop('Habitability_score')
X = home_df
```

### Train-test Split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=12)
```

### I used default parameters for CatBoost and Hyper tuned parameter for Random Forest and XGBoost

##### Fitting Catboost Model

```python
cat = CatBoostRegressor()
```

```python
cat.fit(X_train, y_train, plot=True)
```

##### Fitting Random Forest model

```python
rf = RandomForestRegressor(bootstrap= True,
                           criterion= 'squared_error',
                           max_depth= 80,
                           max_features='auto',
                           min_samples_leaf= 4,
                           min_samples_split=12,
                           n_estimators= 1200 )
```

```python
rf.fit(X_train, y_train)
```

##### Fitting XGBoost model

```python
xg = XGBRegressor(n_estimators=700, learning_rate=0.01, max_depth=10)
```

```python
xg.fit(X_train, y_train)
```

```python
# Function which make prediction on both train and test data and Return train score and test score

def models_score(models):
    for name,model in models.items():
        y_train_pred = model.predict(X_train)
        train_score = r2_score(y_train, y_train_pred)
        
        y_test_pred = model.predict(X_test)
        test_score = r2_score(y_test,y_test_pred)
        
        print(name)
        print("Train Score :{} and Test Score:{}".format(train_score,test_score))
```

```python
# Function which plot graph of residuals

def residual(models):
    
    plt.figure(figsize=(15,15))
    plt.suptitle('Residual Analysis Comparison',fontsize=18)\
    
    a=1
    for name,model in models.items():
        y_train_pred = model.predict(X_train)
        train_residual = y_train-y_train_pred
        
        y_test_pred = model.predict(X_test)
        test_residual = y_test- y_test_pred
        
        plt.subplot(3,2,a)
        plt.title('Train Residual')
        sb.distplot(train_residual, label=name)
        plt.legend()

        plt.subplot(3,2,a+1)
        plt.title('Test Residual')
        sb.distplot(test_residual, label=name)
        plt.legend()
        
        a+=2
    plt.show()
```

```python
models = {'Random Forest':rf, 'XG Boost': xg, 'CatBoost': cat}
```

```python
# calling function

models_score(models)
```

```python
residual(models)
```

### We see XGBoost and Random Forest give almost same results.
